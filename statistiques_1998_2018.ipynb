# Comparaison des performances de l'équipe de france lors des coupes du monde 1998 et 2018

### Import des bibliothèques

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Affichage des dataframes

```python
df_1998=pd.read_csv('data_1998.csv', sep=';')
df_2018=pd.read_csv('data_2018.csv', sep=';')
df_1998
df_2018
```

### Suppression des colonnes inutiles

```python
df_1998 = df_1998.drop("Défense", axis='columns')
df_1998 = df_1998.drop("Coup francs", axis='columns')
df_1998 = df_1998.drop("Corner", axis='columns')
df_2018 = df_2018.drop("Défense", axis='columns')
df_2018 = df_2018.drop("Coup francs", axis='columns')
df_2018 = df_2018.drop("Corner", axis='columns')
```

### Nombre de but total lors des coupes du monde 1998 et 2018

```python
plt.bar(height=[df_1998['Nombre de but'].sum(),df_2018['Nombre de but'].sum()],x=['1998','2018'])
plt.rcParams["figure.figsize"] = (4, 4)
print("Le but d'un joueur adverse contre son camp n'est pas comptabilisé")
plt.ylabel('Comparaison du nombre de buts entre 1998 et 2018')
```

```python
plt.bar(height=df_1998['Nombre de but'], x=df_1998['Nom du joueur'])
plt.xticks(rotation='vertical')
plt.rcParams["figure.figsize"] = (8, 4)
plt.ylabel('Meilleur buteurs de 1998')
```

```python
plt.bar(height=df_2018['Nombre de but'], x=df_2018['Nom du joueur'])
plt.xticks(rotation='vertical')
plt.rcParams["figure.figsize"] = (8, 4)
plt.ylabel('Meilleur buteurs de 2018')
```

### Nombre de passes décisives lors total des coupes du monde 1998 et 2018

```python
plt.bar(height=[df_1998['Nombre de passes décisives'].sum(),df_2018['Nombre de passes décisives'].sum()],x=['1998','2018'])
plt.rcParams["figure.figsize"] = (2, 4)
plt.ylabel('Nombre de passes décisives')
```

### Meilleurs passeurs décisifs en 1998

```python
plt.bar(height=df_1998['Nombre de passes décisives'], x=df_1998['Nom du joueur'])
plt.xticks(rotation='vertical')
plt.rcParams["figure.figsize"] = (8, 4)
plt.ylabel('Joueurs qui ont été impliqués dans le plus de but')
```

### Meilleurs passeurs décisifs en 2018

```python
plt.bar(height=df_2018['Nombre de passes décisives'], x=df_2018['Nom du joueur'])
plt.xticks(rotation='vertical')
plt.rcParams["figure.figsize"] = (8, 4)
plt.ylabel('Joueurs qui ont été impliqués dans le plus de but en 2018')
```

### Nombre de cartons jaunes en total lors des coupes du monde 1998 et 2018

```python
yellow_cards_1998=df_1998['Carton jaune'].sum()
print('Total de cartons jaunes en 1998 :',yellow_cards_1998)
yellow_cards_2018=df_2018['Carton jaune'].sum()
print('Total de cartons jaunes en 2018 :',yellow_cards_2018)
```

```python
plt.bar(height=[df_1998['Arrêts'].sum()/7,df_2018['Arrêts'].sum()/6],x=['Fabien Barthez (1998)','Hugo Lloris (2018)'])
plt.rcParams["figure.figsize"] = (4, 4)
plt.ylabel("Nombre d'arrêts par match")
```
